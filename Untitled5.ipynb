```python
from __future__ import print_function
import cv2 as cv
import re
from os import listdir
import os
from PIL import Image
from numpy import asarray
from skimage import color
from skimage import io
import tensorflow as tf
import numpy as np
```

```python
data_path='uav0000273_00001_v'
```

```python
allfolderss = os.listdir(data_path)
#print(type(allfolderss))
all = []
for i in allfolderss :
   all.append(data_path +'/'+ i )
all.sort(key=lambda f: int(re.sub('\D', '', f)))
print(all)
```

```python
frameseq=[]

for i in all:
  img = io.imread(i)
  #img=tf.image.resize(img, [500,500])
  img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
  img=asarray(img)

  frameseq.append(img)
print(frameseq)
print(type(frameseq))
```

```python
backSub = cv.createBackgroundSubtractorMOG2(history=20,varThreshold=25,detectShadows=False)
size=5
#size(5,5)
for j in range(len(frameseq)):
    fgMask = backSub.apply(frameseq[j])
    img = cv.medianBlur(img, size)
    #img = cv.GaussianBlur(img, size)
    
    cv.imshow("frame",frameseq[j])
    cv.imshow("mask",fgMask)
    
    key=cv.waitKey(30)
    if key== 27:
        break
    
```
